```python
data_test = pd.read_csv('test.csv')
```

```python
data_test
```

```python
len(data_test)
```

```python
import numpy as np
# 행 = 418, 열 = 2 (PassengerId, Survived)
answer = np.zeros([len(data_test), 2]) # 0으로 초기화된 2차원 배열 생성

# 생존 여부 규칙 예측
# 1. 1등석 승객은 생존한다.
# 2. 2등석 승객 중 여성은 생존한다.
# -> answer는 이미 0으로 모두 초기화하였으므로,
#    1(생존)으로 바꾸는 경우만 지정하면 됨!
```

```python
answer[:,0] = data_test['PassengerId']
```

```python
answer
```

```python
data_test['Pclass']==1
```

```python
# Pclass행의 1열이 모두 1(=생존)이라고 기록
answer[data_test['Pclass']==1, 1] = 1
```

```python
answer
```

```python
# 2등석이고, 여성인 행의 1열을 1(=생존)
answer[(data_test['Pclass']==2) & (data_test['Sex']=='female'), 1] = 1
```

```python
answer
```

```python
type(answer)
# numpy 배열에서 csv 파일로 바로 저장할 수 X
```

```python
# csv로의 저장을 위해 pandas 라이브러리의 DataFrame 형식으로 변환하여 저장
df_answer = pd.DataFrame(answer)
```

```python
df_answer
```

```python
# 각 열의 제목 지정
df_answer.columns = ['PassengerId', 'Survived']
```

```python
# 제출형식(Integer) 변경
# 실수형으로 되어 있는 배열의 값 -> 정수형
df_answer = df_answer.astype('int32') # 4Byte
```

```python
df_answer
```

```python
df_answer.to_csv('test_predict.csv',index = False)
```
